```python
# ============================================================
# SHAP 
# Uses Hopsworks Feature Store + Model Registry
# ============================================================

import os
import shap
import hopsworks
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import joblib

# ------------------------------------------------------------
# Login to Hopsworks
# ------------------------------------------------------------
project = hopsworks.login(
    api_key_value=os.getenv("HOPSWORKS_API_KEY"),
    project=os.getenv("HOPSWORKS_PROJECT_NAME")
)

fs = project.get_feature_store()
mr = project.get_model_registry()

# ------------------------------------------------------------
# Load Feature Group (NO targets)
# ------------------------------------------------------------
fg = fs.get_feature_group("aqi_karachi_features_final", version=3)
df = fg.read().sort_values("timestamp").reset_index(drop=True)

DROP_COLS = ["timestamp", "aqi", "target_day1", "target_day2", "target_day3"]
if "season" in df.columns:
    DROP_COLS.append("season")

X = df.drop(columns=[c for c in DROP_COLS if c in df.columns])
X = X.fillna(X.median())

# SHAP sample
X_sample = X.sample(min(500, len(X)), random_state=42)
print(f"âœ… SHAP sample ready: {X_sample.shape}")

# ------------------------------------------------------------
# Helper: Load actual trained model
# ------------------------------------------------------------
def load_hopsworks_model(mr, name, version=1):
    model_meta = mr.get_model(name, version=version)
    model_dir = model_meta.download()
    model_path = os.path.join(model_dir, "model.pkl")
    return joblib.load(model_path)

# ------------------------------------------------------------
# Load models 
# ------------------------------------------------------------
best_models = {
    "Day1": {
        "Ridge": load_hopsworks_model(mr, "aqi_ridge_day1"),
        "RF":    load_hopsworks_model(mr, "aqi_randomforest_day1"),
        "XGB":   load_hopsworks_model(mr, "aqi_xgboost_day1"),
    },
    "Day2": {
        "Ridge": load_hopsworks_model(mr, "aqi_ridge_day2"),
        "RF":    load_hopsworks_model(mr, "aqi_randomforest_day2"),
        "XGB":   load_hopsworks_model(mr, "aqi_xgboost_day2"),
    },
    "Day3": {
        "Ridge": load_hopsworks_model(mr, "aqi_ridge_day3"),
        "RF":    load_hopsworks_model(mr, "aqi_randomforest_day3"),
        "XGB":   load_hopsworks_model(mr, "aqi_xgboost_day3"),
    }
}

# ------------------------------------------------------------
# Helper: Top SHAP features
# ------------------------------------------------------------
def get_top_features(shap_values, feature_names, top_n=15):
    mean_abs = np.abs(shap_values).mean(axis=0)
    return pd.DataFrame({
        "feature": feature_names,
        "importance": mean_abs
    }).sort_values("importance", ascending=False).head(top_n)

# ------------------------------------------------------------
# Output folder
# ------------------------------------------------------------
os.makedirs("shap_plots", exist_ok=True)

# ------------------------------------------------------------
# SHAP ANALYSIS + COMPARISON
# ------------------------------------------------------------
for day, models in best_models.items():
    print(f"\n================ {day} =================")
    top_features = {}

    for model_name, model in models.items():
        print(f"ðŸ” SHAP for {model_name}")

        if model_name == "Ridge":
            X_used = X_sample.iloc[:100]
            explainer = shap.KernelExplainer(model.predict, X_used)
            shap_values = explainer.shap_values(X_used)
        else:
            X_used = X_sample
            explainer = shap.TreeExplainer(model)
            shap_values = explainer.shap_values(X_used)

        top_features[model_name] = get_top_features(
            shap_values, X_used.columns
        )

        # -------- Bar plot --------
        plt.figure(figsize=(10,6))
        shap.summary_plot(
            shap_values, X_used,
            plot_type="bar",
            max_display=15,
            show=False
        )
        plt.title(f"{model_name} - {day} Feature Importance")
        plt.tight_layout()
        plt.savefig(f"shap_plots/{day}_{model_name}_bar.png", dpi=300)
        plt.close()

        # -------- Beeswarm --------
        plt.figure(figsize=(10,8))
        shap.summary_plot(
            shap_values, X_used,
            max_display=15,
            show=False
        )
        plt.title(f"{model_name} - {day} Feature Impact")
        plt.tight_layout()
        plt.savefig(f"shap_plots/{day}_{model_name}_beeswarm.png", dpi=300)
        plt.close()

    # --------------------------------------------------------
    # Side-by-side comparison
    # --------------------------------------------------------
    fig, axes = plt.subplots(1, 3, figsize=(20,7))
    colors = ["steelblue", "forestgreen", "darkorange"]

    for i, model_name in enumerate(models.keys()):
        df_plot = top_features[model_name]
        axes[i].barh(df_plot["feature"], df_plot["importance"], color=colors[i])
        axes[i].invert_yaxis()
        axes[i].set_title(f"{model_name} - {day}")
        axes[i].grid(axis="x", alpha=0.3)

    plt.suptitle(
        f"{day} Model Comparison (SHAP Feature Importance)",
        fontsize=18
    )
    plt.tight_layout()
    plt.savefig(f"shap_plots/{day}_comparison.png", dpi=300)
    plt.show()
```
